```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading in data
march2020 = pd.read_csv('./march2020.csv')
may2020 = pd.read_csv('./may2020.csv')
july2020 = pd.read_csv('./july2020.csv')

# Removing commas and converting to float for all of the string elements
def parse(df, col):
    clean = []
    for elem in df[col].values:
        elem = elem.replace(',', '')
        clean.append(float(elem))
        
    df[col] = clean
    
colList = ['totalSoybeanMealSupply', 'totalSoybeanMealDemand', 'soybeanOilSupply', 'soybeanOilDemand']

def parseAllCol(df, listOfCol):
    for col in listOfCol:
        parse(df, col)
        
parseAllCol(march2020, colList)

parseAllCol(may2020, colList)

parseAllCol(july2020, colList)
```

```python
corrMarch = march2020.corr()
mar = sns.heatmap(corrMarch, vmax = 1.0, square = True)
mar.set_title('March 2020 Contract Date Feature and Label Correlation')
```

```python
corrMay = may2020.corr()
may = sns.heatmap(corrMay, vmax = 1.0, square = True)
may.set_title('May 2020 Contract  Date Feature and Label Correlation')
```

```python
corrJuly = july2020.corr()
jul = sns.heatmap(corrJuly, vmax = 1.0, square = True)
jul.set_title('July 2020 Contract Date Feature and Label Correlation')
```

```python
# Writing to csv
corrMarch.to_csv('./corrMarch.csv')
corrMay.to_csv('./corrMay.csv')
corrJuly.to_csv('./corrJuly.csv')
```
